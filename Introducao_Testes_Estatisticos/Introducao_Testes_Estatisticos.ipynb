### ZTest para uma amostra:

```python
filmes = pd.read_csv('dados/movies.csv')
```

```python
filmes.query('movieId==1')
```

```python
notas1 = notas.query('movieId==1')
notas1.head()
```

```python
ax = sns.distplot(notas1['rating'])
ax.set(xlabel='Nota', ylabel='Densidade')
ax.set_title('Distribuição das notdas do filme Toy Story')
```

```python
ax = sns.boxplot(x=notas1['rating'])
ax.set(xlabel='Notas')
ax.set_title('Distribuição das notas do filme Toy Story', fontsize=16)
```

```python
notas1.rating.count()
```

```python
# Sample com número suficiente posso aplicar o intervalo de confiança com estatítica z:
zconfint(notas1.rating)
```

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating, value=3.4320503405352603)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value=3.4320503405352603)
    return(i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])


fig = plt.figure(figsize=(15,8))
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y=0.05, xmin=2, xmax=len(temp), colors='r')
```

### Comparação de dois conjuntos de amostras:

```python
print(ztest(notas1.rating, notas.rating))
zconfint(notas1.rating, notas.rating) # intervalo do quão diferente é a nota do filme Toy Story dos outros filmes da amostra.
```

```python
print(ztest(notas.rating, notas1.rating))
zconfint(notas.rating, notas1.rating) # intervalo do quão diferente é a nota dos filmes da amostra em comparação com o Toy Story.
```

```python
from scipy.stats import ttest_ind

ttest_ind(notas.rating, notas1.rating)
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toystory = DescrStatsW(notas1.rating)
```

```python
comparacao = descr_todas_as_notas.get_compare(descr_toystory)
comparacao.summary() #use_t por padrão é True
```

```python
comparacao.summary(use_t=False)
```
