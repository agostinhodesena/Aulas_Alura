# Data Visualization - Matplotlib

### Fazendo as primeiras visualizações:

```python
# Importação de bibliotecas:
import pandas as pd
import datetime
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importação da base de dados:
df = pd.read_csv('dados/monitoramento_tempo.csv')
```

```python
df.head()
```

```python
# Informações sobre o DF:
df.info()
```

```python
# Convertendo a coluna data em um objeto date-time:
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.figure(figsize=(15,8)) # Tamanho do gráfico
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

### Funcionalidades básicas do matplotlib

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1]) # Dando eixos com espaço 100% ao gráfico.
eixo.plot(df['data'], df['temperatura'], color='green')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15) #loc = posição
```

### Visualização sobre visualização em um mesmo gráfico

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.05,0.7,0.25,0.25]) # Visualização menor 

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) # gráfico de apenas um trecho da série.
eixo.set_title('Temperatura em 2014', fontsize=25, pad=20) # pad = espaçamento entre título e gráfico.
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15) # loc = best, matplotlib escolhe o melhor lugar

eixo2.plot(df['data'], df['temperatura'], color = 'b')
eixo2.set_title('Temperatura no momento', fontsize=15) 
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.legend(['Temperatura'], loc='lower right', fontsize=8) # loc = best, matplotlib escolhe o melhor lugar
eixo2.grid(True)
```

### Uma visualização mais complexa

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.05,0.7,0.25,0.25]) # Visualização menor 

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1)) # gráfico de apenas um trecho da série.
eixo.set_ylim(270,320) # limitando eixo y.
eixo.set_title('Temperatura em maio de 2014', fontsize=25, pad=20) # pad = espaçamento entre título e gráfico.
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15) # loc = best, matplotlib escolhe o melhor lugar

azul_esquerda = df['data'] < datetime.datetime(2014,5,1)
azul_direita = df['data'] > datetime.datetime(2014,6,1)

eixo2.plot(df['data'], df['temperatura'], color = 'g')
eixo2.plot(df[azul_esquerda]['data'], df[azul_esquerda]['temperatura'], color='b')
eixo2.plot(df[azul_direita]['data'], df[azul_direita]['temperatura'], color='b')
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) # gráfico de apenas um trecho da série.
eixo2.set_title('Temperatura em 2014', fontsize=15) 
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.legend(['Temperatura'], loc='lower right', fontsize=8) # loc = best, matplotlib escolhe o melhor lugar
eixo2.grid(True)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

# Indicando os locais onde a anotação deve ficar:
x1 = df['data'][df['temperatura'].idxmax()] # idxmax = index máximo
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000] # idxmax = index máximo
y2 = max(df['temperatura']) - 5 

# Anotação e seta:
eixo.annotate("Máximo", xy=(x1,y1), fontsize=16, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

# Indicando os locais onde a anotação deve ficar:
x1 = df['data'][df['temperatura'].idxmin()] # idxmin = index mínimo
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 7000] # idxmin = index máximo
y2 = min(df['temperatura']) + 5 

# Anotação e seta:
eixo.annotate("Mínimo", xy=(x1,y1), fontsize=16, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

eixo.axhline(max(df['temperatura']), color='k', linestyle='--') # Inserindo linhas horizontais no gráfico. Max() = fórmula para valor máximo. Color k = preto. Linestyle -- = tracejado
eixo.axhline(min(df['temperatura']), color='k', linestyle='--') # Inserindo linhas horizontais no gráfico. Min() = fórmula para valor máximo. Color k = preto. Linestyle -- = tracejado
```

### Algumas outras visualizações

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='black')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias) # Colocando o rótulo com o nome dos dias na tabela.
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

# Colorindo os dados e usando marcadores específicos para cada espécie:
cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'} # x = xzinho, o = bolinha, v = triangulinho.

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                 color=cores[especie], marker=marcadores[especie],
                 s=100)

eixo.set_title('Gráfico de dispersão', fontsize=25, pad=10)
eixo.set_xlabel('Comprimento da Sépala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)
eixo.tick_params(labelsize=15) # Aumentando o tamanho da fonte dos números que são o rótulo do gráfico
eixo.legend(cores, fontsize=20)
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

# Colocando cores nas caixas do boxplot:
cores = ['red', 'blue', 'orange', 'green']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True) # axis=1 - retirar coluna "espécie", patch_artist=True - colorindo as caixas, 
eixo.set_title('Boxplot', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

# Colocando cores nas caixas do boxplot:
for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)

# Colocando os valores outliers como marcadores 'x':
for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```

### Histograma

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, orientation='vertical', density=True) # Mudando a orientação: orientation='horizontal'; Mudando a densidade: density=True
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

### Histograma com média, mediana...

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# Obtendo a média e a variância da variável:
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20) # Mudando a orientação: orientation='horizontal'; Mudando a densidade: density=True
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

# Inserindo a média e a variância no gráfico:
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu,sigma),
             xy=(5.7, 20), fontsize=15)

# Inserindo uma reta para a média:
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1.3, 28), fontsize=20)

eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='--')
eixo.annotate('Mediana', xy=(df['comprimento_pétala'].median()+0.2, 31), fontsize=20)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# Selecionando apenas a espécie "Iris-versicolor"
df_iv = df[df['espécie'] == 'Iris-versicolor']

# Obtendo a média e a variância da variável:
mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20) # Mudando a orientação: orientation='horizontal'; Mudando a densidade: density=True
eixo.set_title('Histograma - Iris Versicolor', fontsize=25, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

# Inserindo a média e a variância no gráfico:
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu,sigma),
             xy=(4.6, 6), fontsize=15)

# Inserindo uma reta para a média:
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-0.5, 5.5), fontsize=20)

eixo.axvline(df_iv['comprimento_pétala'].median(), color='g', linestyle='--')
eixo.annotate('Mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=20,
             color='g')

fig.savefig('imagens/histograma_iv.png', bbox_inches='tight') #tight - salvando imagem com 'borda branca' pequena
```

### Juntando imagens

```python
largura, altura = Image.open('imagens/histograma_ivg.png').size
combinada = Image.new("RGB", (3*largura, altura))
intervalo = 0

for imagem in map(Image.open, ['imagens/histograma_iv.png', 'imagens/histograma_is.png', 'imagens/histograma_ivg.png']):
    combinada.paste(imagem, (intervalo,0))
    intervalo += largura
    
combinada.save('imagens/combinada.png')
```
